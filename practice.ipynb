```python
# COnverting to GRAY
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_gray.shape
```

```python
cv2.imshow('Image',img_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(img_gray, cmap = 'gray')
```

```python
th, img_bin = cv2.threshold(img_gray, 120,255, cv2.THRESH_BINARY)
img_bin.shape
```
